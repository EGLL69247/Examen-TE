### Hacemos la tabla de frecs

```python
pd.set_option('precision',6)
datos= med.sort_values('Medida')
bins= [-np.inf,datos,np.inf]
k= 1+3.332*m.log10(len(datos))
numero= int(k)
if numero % 2==0:
    periodos= m.ceil(k)
else:
    periodos= int(k)
    
inf= float(datos.min())   
dif= float(datos.max())
sup= float(datos.max())

interv= pd.interval_range(
    start= inf,
    end= sup,
    periods= k,
    name= 'Intervalo',
    closed= 'left')

df= pd.DataFrame(index= interv)
df['FrecAbs']= df.groupby(med['Medida'].tolist(),as_index=False).size()
df['Marca']= df.index.mid
df
```
